```python
import pandas as pd
```

```python
data_adidas_df = pd.read_csv('../Data/adidas.csv')
data_adidas_df
```

```python
import tiktoken

# FUncion para contar el numero de tokens en un texto
def num_tokens_from_string(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
# Funcion para crear los documentos a partir de la data del dataset
def create_document_from_adidas_data(name, color, categories, description):
    
    document = f"Name: {name}\nColor: {color}\nCategories: {categories}\nDescription: {description}"
    
    return document
```

```python
data_adidas_df['cloth_document'] = data_adidas_df.apply(lambda x: create_document_from_adidas_data(x['name'], x['color'], x['breadcrumbs'], x['description']), axis=1)
```

```python
data_adidas_df
```

```python
data_adidas_df['cloth_document_token_count'] = data_adidas_df.apply(lambda x: num_tokens_from_string(x['cloth_document'], 'cl100k_base'), axis=1)
```

```python
data_adidas_df.hist(column='cloth_document_token_count', bins=30)
```

```python
data_adidas_df_json = data_adidas_df[['name', 'color','breadcrumbs','description','cloth_document']]
data_adidas_df_json
```
